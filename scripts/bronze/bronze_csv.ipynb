```python
#Define ingestion config

INGESTION_CONFIG=[{
    "path": "/Volumes/workspace/ecommerce_bronze/raw_sources/df_Customers.csv",
    "table": "df_Customers"
},{
    "path": "/Volumes/workspace/ecommerce_bronze/raw_sources/df_OrderItems.csv",
    "table": "df_OrderItems"
},{
    "path":"/Volumes/workspace/ecommerce_bronze/raw_sources/df_Orders.csv",
    "table": "df_Orders"
},{
    "path": "/Volumes/workspace/ecommerce_bronze/raw_sources/df_Payments.csv",
    "table": "df_Payments"
},{
    "path": "/Volumes/workspace/ecommerce_bronze/raw_sources/df_Products.csv",
    "table": "df_Products"
}]
```

```python
#reading and writing data

for i in INGESTION_CONFIG:
    df=(spark.read.option("header",True).option("inferSchema",True).csv(i["path"]))
    df.write.mode("overwrite").format("delta").saveAsTable(f"workspace.ecommerce_bronze.{i['table']}")
    print(f"Table {i['table']} is Created succesfully!")
```
