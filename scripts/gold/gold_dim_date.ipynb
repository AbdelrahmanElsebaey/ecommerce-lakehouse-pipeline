```python
query="""
  WITH all_dates AS (

    SELECT DATE(order_purchase_timestamp) AS date FROM workspace.ecommerce_silver.df_orders
    WHERE order_purchase_timestamp IS NOT NULL
    UNION
    SELECT DATE(order_approved_at) FROM workspace.ecommerce_silver.df_orders
    WHERE order_approved_at IS NOT NULL
    UNION
    SELECT DATE(order_delivered_timestamp) FROM workspace.ecommerce_silver.df_orders
    WHERE order_delivered_timestamp IS NOT NULL
    UNION
    SELECT DATE(order_estimated_delivery_date) FROM workspace.ecommerce_silver.df_orders
    WHERE order_estimated_delivery_date IS NOT NULL

)

SELECT DISTINCT
    date,
    DAY(date) AS day,
    MONTH(date) AS month,
    YEAR(date) AS year,
    QUARTER(date) AS quarter,
    DATE_FORMAT(date, 'EEEE') AS day_name,
    DATE_FORMAT(date, 'MMMM') AS month_name

FROM all_dates
ORDER BY date;
"""
df=spark.sql(query)
```

```python
display(df.limit(10))
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.ecommerce_gold.dim_date")
```
