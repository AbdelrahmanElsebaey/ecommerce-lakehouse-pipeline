```python
query="""
SELECT 
  orders.order_id,
  orders.customer_id,
  DATE(orders.order_purchase_timestamp) AS order_date,
  DATE(orders.order_delivered_timestamp) AS order_delivered_date,
  orders.delivery_delay_days,
  items.product_id,
  items.seller_id,
  items.item_total_value as revenue,
  items.payment_value_with_tax as total_payment
FROM workspace.ecommerce_silver.df_orders as orders
INNER JOIN workspace.ecommerce_silver.df_orderitems as items
ON orders.order_id = items.order_id
"""
df = spark.sql(query)
```

```python
display(df.limit(5))
```

```python
df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").saveAsTable("workspace.ecommerce_gold.fact_orders")
```
